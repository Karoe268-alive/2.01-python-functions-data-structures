```python
def validate_records(records):
    if not isinstance(records, list):
        raise TypeError(f"Expected a list, got {type(records).__name__}")
    
    if len(records) == 0:
        raise ValueError("Records list is empty")
    
    for i, item in enumerate(records):
        if not isinstance(item, dict):
            raise TypeError(f"Item at index {i} must be a dictionary, got {type(item).__name__}")
```

```python
def clean_names(records):
    return [record for record in records if record["name"] != ""]
```

```python
records = [
    {"name": "Ava", "score": 95, "hours_studied": 10},
    {"name": "Ben", "score": None, "hours_studied": 8},
    {"name": "Cara", "score": 88, "hours_studied": None},
    {"name": "", "score": 72, "hours_studied": 6},      
    {"name": "Eli", "score": 105, "hours_studied": 9},
    {"name": "Finn", "score": 65, "hours_studied": 7}
]

cleaned = clean_names(records)

for r in cleaned:
    print(r)
```

```python
def fill_missing_values(records):
    valid_scores = [r["score"] for r in records if r["score"] is not None]
    avg_score = sum(valid_scores) / len(valid_scores)

    valid_hours = [r["hours_studied"] for r in records if r["hours_studied"] is not None]
    avg_hours = sum(valid_hours) / len(valid_hours)

    for record in records:
        if record["score"] is None:
            record["score"] = round(avg_score, 1)
        if record["hours_studied"] is None:
            record["hours_studied"] = round(avg_hours, 1)

    return records
```

```python
records = [
    {"name": "Ava",  "score": 95,  "hours_studied": 10},
    {"name": "Ben",  "score": None,"hours_studied": 8},   
    {"name": "Cara", "score": 88,  "hours_studied": None},
    {"name": "Finn", "score": 65,  "hours_studied": 7}
]

filled = fill_missing_values(records)

for r in filled:
    print(r)
```

```python
def cap_scores(records):
    for record in records:
        if record["score"] is not None:
            if record["score"] > 100:
                record["score"] = 100
            elif record["score"] < 0:
                record["score"] = 0
    return records
```

```python
records = [
    {"name": "Ava",  "score": 95},
    {"name": "Eli",  "score": 105},  
    {"name": "Finn", "score": -10},  
    {"name": "Ben",  "score": None}, 
    {"name": "Cara", "score": 100},  
]

capped = cap_scores(records)

for r in capped:
    print(r)
```

```python
import pandas as pd

def to_dataframe(records):
    return pd.DataFrame(records)
```

```python
records = [
    {"name": "Ava",  "score": 95,   "hours_studied": 10},
    {"name": "Ben",  "score": 82.7, "hours_studied": 8},
    {"name": "Cara", "score": 88,   "hours_studied": 8.3},
    {"name": "Finn", "score": 65,   "hours_studied": 7}
]

df = to_dataframe(records)
print(df)
print(df.dtypes)
```

```python
def clean_dataset(records):
    validate_records(records)           
    records = clean_names(records)      
    records = fill_missing_values(records)  
    records = cap_scores(records)       
    df = to_dataframe(records)          
    return df
```

```python
records = [
    {"name": "Ava",  "score": 95,  "hours_studied": 10},
    {"name": "Ben",  "score": None,"hours_studied": 8},
    {"name": "Cara", "score": 88,  "hours_studied": None},
    {"name": "",     "score": 72,  "hours_studied": 6},
    {"name": "Eli",  "score": 105, "hours_studied": 9},
    {"name": "Finn", "score": 65,  "hours_studied": 7}
]

df = clean_dataset(records)
print(df)
```
